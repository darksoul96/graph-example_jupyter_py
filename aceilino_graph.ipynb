```python
import pandas
```

```python
from matplotlib import pyplot
```

```python
import numpy
```

```python
colnames=['X', 'Y'] 
data = pandas.read_table('ACEILINO.TBL', names=colnames, delimiter=' ', header=None)
```

```python
data.X
```

```python
data.Y
```

```python
pyplot.plot(data.X, data.Y)
pyplot.plot(data.X, data.Y + 2)

# Find the index of the maximum value in y
max_index = numpy.argmax(data.Y)

# Get the x and y values at the maximum index
max_x = data.X[max_index]
max_y = data.Y[max_index]

pyplot.annotate(f'Maximum value: {max_y:.2f}', xy=(max_x, max_y), xytext=(max_x - 0.1, max_y - 0.1))
pyplot.title("Aceilino")
pyplot.xlabel("X")
pyplot.ylabel("Y")
pyplot.show()
```
